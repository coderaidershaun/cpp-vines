# Fit a Clayton copula

This example loads the six asset price series from `prices.csv`, converts them to empirical marginals, then fits a Clayton copula to BTC and ETH.

```python
from helpers import ASSET_NAMES, PRICES_PATH, load_assets
from vines import Clayton, Result
```

```python
# Load asset price data
assets = load_assets()
assert len(assets) == 6
print(f"loaded {len(assets[0].prices())} prices for {', '.join(ASSET_NAMES)} from {PRICES_PATH}")
```

```python
u1 = assets[0].u_values()
u2 = assets[1].u_values()
copula = Clayton(u1, u2)
fit = copula.fit()

assert fit.result is Result.SUCCESS
print(f"fitted parameter bounds: {copula.params}")
print(f"AIC: {fit.aic:.2f}")
```

```python
point = (u1[-1], u2[-1])
conditionals = copula.h_conditional_prob_set(*point)

print(f"density: {copula.estimate_copula_density(*point):.2f}")
print(f"u1 given u2: {conditionals.u1_given_u2:.2f}")
print(f"u2 given u1: {conditionals.u2_given_u1:.2f}")
print(f"sample log-likelihood: {copula.estimate_copula_prob_densities().ln_likelihood:.2f}")
```
