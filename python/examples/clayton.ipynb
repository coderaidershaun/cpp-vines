# Fit a Clayton copula

This example fits one copula directly, then evaluates its density and conditional probabilities.

```python
from vines import Clayton, Result
```

```python
u1 = [0.08, 0.16, 0.27, 0.39, 0.51, 0.64, 0.73, 0.88]
u2 = [0.11, 0.23, 0.19, 0.46, 0.42, 0.69, 0.61, 0.84]

initial_alpha = Clayton.alpha_from_kendalls_tau(0.35)
copula = Clayton(u1, u2, alpha_init=initial_alpha)
fit = copula.fit()

assert fit.result is Result.SUCCESS
print("fitted parameter bounds:", copula.params)
print("AIC:", fit.aic)
```

```python
point = (0.35, 0.55)
conditionals = copula.h_conditional_prob_set(*point)

print("density:", copula.estimate_copula_density(*point))
print("u1 given u2:", conditionals.u1_given_u2)
print("u2 given u1:", conditionals.u2_given_u1)
print("sample log-likelihood:", copula.estimate_copula_prob_densities().ln_likelihood)
```
